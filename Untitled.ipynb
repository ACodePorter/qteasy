```python
import tqdm
from tqdm import tqdm, trange
import time
```

```python

for _ in trange(1000, desc='This is a quite long description'):
    time.sleep(0.02)
        
```

```python
outer_iterable = range(5)
inner_iterable = lambda i: range(100 * (i+1))  # inner length varies for demo 
for i in tqdm(outer_iterable, desc="Outer loop"):
    for j in tqdm(inner_iterable(i), desc=f"Inner loop {i}", position=1, leave=False):
        # do some work
        time.sleep(0.01)
```

```python
for i in trange(3, desc='Outer loop', leave=True):
    for j in trange(5, desc='Inner loop', leave=False):
        for j in trange(4, desc='Inner inner loop', leave=False):
            time.sleep(0.05)
        
```
