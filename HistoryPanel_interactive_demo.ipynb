# HistoryPanel 交互式图表示例

本 Notebook 演示如何在 Jupyter 环境中使用 `HistoryPanel.plot(interactive=True)` 绘制交互式 K 线图。

```python
import os
import sys

# 确保可以在 notebook 中导入本地安装的 qteasy
sys.path.insert(0, os.path.abspath('..'))

import qteasy as qt

# 获取演示用的 HistoryPanel
hp = qt.get_history_data(
    htypes='open, high, low, close, vol',
    shares='000300.SH',
    start='20230101',
    end='20231231',
)
print(hp)

# 静态 K 线 + 成交量
hp.plot(interactive=False)

# 交互式 K 线（在 Jupyter + anywidget/ipywidgets 环境中体验更佳）
fig = hp.plot(interactive=True)
fig
```
